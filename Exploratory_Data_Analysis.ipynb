```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://github.com/hagarwal011/Movie_Data_Analysis/raw/refs/heads/main/Movie_data.csv")
```

```python
df
```

```python
df = df.drop(columns = ['Unnamed: 0'])
```

```python
df.columns = ['Title', 'Rating', 'Release Date', 'Runtime', 'Genres']
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['Release Year'] = pd.to_datetime(df['Release Date']).dt.year
```

```python
df
```

```python
df['Release Year'].value_counts()
```

```python
df['Rating'] = df['Rating'].round(1)
```

```python
df['Release Year'] = df['Release Year'].astype(int)
```

```python
df = df[df['Release Year'] < 2025]
```

```python
df['Release Month'] = pd.to_datetime(df['Release Date']).dt.month_name()
```

```python
df
```

```python
df['Release Weekday'] = pd.to_datetime(df['Release Date']).dt.day_name()
```

```python
df
```

```python
df.to_csv('Movie_data_cleaned.csv')
```
